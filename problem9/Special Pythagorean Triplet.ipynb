```python
"""

A Pythagorean triplet is a set of three natural numbers, a < b < c, for which, a^2 + b^2 = c^2

For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.

Find the product abc.

"""
```

```python
# POC

"""
# constraints
1. a < b < c
2. a + b + c = 1000
3. a^2 + b^2 = c^2

# remove a
let a = 1000 - b - c
so a^2 
= (1000 - b - c)^2 = (1000 - b - c)*(1000 - b - c) 
= (1000000 - 1000b - 1000c) + (-1000b + b^2 + bc) + (-1000c + cb + c^2) 
= 1000000 - 2000b - 2000c + 2bc + b^2 + c^2

so a^2 + b^2 = c^2
becomes 1000000 - 2000b - 2000c + 2bc + b^2 + c^2 + b^2 = c^2

# reduce
1000000 - 2000b - 2000c + 2bc + b^2 + c^2 + b^2 = c^2
1000000 - 2000b - 2000c + 2bc + 2b^2 = 0
1000000 - 2000b - (2000 - 2b)c + 2b^2 = 0
1000000 - 2000b  + 2b^2 = (2000 - 2b)c
(1000000 - 2000b  + 2b^2)/(2000 - 2b) = c
(500000 - 1000b  + b^2)/(1000 - b) = c

soooooo
a = 1000 - b - c
c = (500000 - 1000b  + b^2)/(1000 - b)

"""


def find_pythagorean(num):
    for b in range(num):
        c = (500000 - 1000*b  + b**2)/(1000 - b)
        a = 1000 - b - c
        #print(a,b,c)
        if a%1 == 0 and c%1 == 0 and a >=0 and c >= 0 and a<b<c:
            return int(a*b*c)
            

print(find_pythagorean(1000))
# solved, but not efficient cuz scanning all b
```
