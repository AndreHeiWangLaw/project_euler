```python
# POC

num = 1
low = 0
high = 10
l = True

while l:
    num += 1
    counter = 0
        
    for i in range(high, low, -1):
        # print(num, i, num%i, counter)
        if num%i == 0:
            counter += 1
            if counter == high:
                print(num)
                l = False
                break

# also correct, but no clean soln
```

```python
# Least Common Multiple (LCM) with prime factorisation

low = 1
high = 10
lnum2 = lnum3 = lnum5 = lnum7 = lnum11 = lnum13 = lnum17 = lnum19 = 0


for i in range(low, high+1):
    num2 = num3 = num5 = num7 = num11 = num13 = num17 = num19 = 0
    
    # even
    while i%2 == 0:
        num2 += 1
        i /= 2

    while i%3 == 0:
        num3 += 1
        i /= 3
        
    while i%5 == 0:
        num5 += 1
        i /= 5
        
    while i%7 == 0:
        num7 += 1
        i /= 7
        
    while i%11 == 0:
        num11 += 1
        i /= 11
        
    while i%13 == 0:
        num13 += 1
        i /= 13
        
    while i%17 == 0:
        num17 += 1
        i /= 17
        
    while i%19 == 0:
        num19 += 1
        i /= 19

    lnum2 = max(lnum2, num2)
    lnum3 = max(lnum3, num3)
    lnum5 = max(lnum5, num5)
    lnum7 = max(lnum7, num7)
    lnum11 = max(lnum11, num11)
    lnum13 = max(lnum13, num13)
    lnum17 = max(lnum17, num17)
    lnum19 = max(lnum19, num19)

        
    
answer = 2**lnum2 * 3**lnum3 * 5**lnum5 * 7**lnum7 * 11**lnum11 * 13**lnum13 * 17**lnum17 * 19**lnum19 
print(answer)
```

```python
# optimizing

def get_lcm(low, high):
    prime_expo = {}
    

    for i in range(low, high+1):
        # prince factorisation
        d = 2
        current_n = i
        
        while d*d <= current_n:
            if current_n%d == 0:
                count = 0
                while current_n%d == 0:
                    count += 1
                    current_n /= d
                prime_expo[d] = max(prime_expo.get(d, 0), count)
            d += 1

        # prime
        if current_n > 1:
            prime_expo[current_n] = max(prime_expo.get(current_n, 0), 1)

    answer = 1
    for key, val in prime_expo.items():
        answer *= key**val
        
    return answer

print(get_lcm(1,20))
```
