```python
# POC

num = 1
for i in range(1,11):
    num = num*i
    print(i, num)

print(num)
# this is likely also correct, just not smallest
```

```python
# POC

num = 1
low = 0
high = 10
l = True

while l:
    num += 1
    counter = 0
        
    for i in range(high, low, -1):
        # print(num, i, num%i, counter)
        if num%i == 0:
            counter += 1
            if counter == high:
                print(num, num, num)
                l = False
                break

    
```
