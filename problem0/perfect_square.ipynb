```python
""" 
[Problem 0]

A number is a perfect square, or a square number, if it is the square of a positive integer.
For example, 25 is a square number because 5^2=5×5=25; it is also an odd square.

The first 5 square numbers are: 1, 4, 9, 16, 25, and the sum of the odd squares is 1+9+25=35.

Among the first 751 thousand square numbers, what is the sum of all the odd squares?

"""
```

```python
# POC (small test)

for i in range(10):
    x = i**2
    if x%2 != 0:
        print(x) 
```

```python
# scale to 751k

answer = 0
the_firth_x = 751000

for i in range(the_firth_x):
    x = i**2
    if x%2 != 0:
        answer = answer + x

print(answer)
```
