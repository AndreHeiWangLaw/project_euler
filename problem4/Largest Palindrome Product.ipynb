```python
# POC

# make a function to check if palindromic number or not
def check_palindromic(num):
    s = str(num)
    index = 0
    r = len(s)
    for i in range(r//2):
        if s[i] != s[r-1-i]:
            return False
    return True

# loop to try all combo



print(check_palindromic(22322))

```
