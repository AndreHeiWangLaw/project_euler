```python
"""
[Problem 1]

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of 
these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

"""
```

```python
# POC (only mult 3)

num = 3
belowX = 10

mult = 1
answer = 0

while 0 < belowX:
    x = num*mult
    if x < belowX:
        answer = answer + x
    else:
        break
    mult += 1
    

print(answer)
```
