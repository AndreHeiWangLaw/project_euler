```python
"""

The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

"""
```

```python
# POC

def primes_under_num(num):

    # 1. Create dict from 1 to num where they are all True
    primes = {}
    for i in range(2, num+1):
        primes[i] = True

    # 2. False for all non-prime
    for i in primes:
        if primes[i] == True:
            # is a primme
            mult = 2
            while i*mult <= num:
                primes[i*mult] = False
                mult += 1

    answer = 0
    for i in primes:
        if primes[i] == True:
            answer += i
    return answer # if 1 is prime, dont forget to +1 for sum

print(primes_under_num(10))
```
