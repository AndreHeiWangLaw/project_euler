```python
"""

The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143?

"""
```

```python
# POC trail division method

num = 70

# step 1: make it odd
while num%2 == 0: # is dividable by 2
    num /= 2

# step 2: test odd 
while num%3 == 0: 
    num /= 3
while num%5 == 0: 
    num /= 5
while num%7 == 0: 
    num /= 7

print(num)
# largest factor is 7
```

```python
# POC trail division method - clean up

num = 70

# step 1: make it odd
while num%2 == 0: # is dividable by 2
    num /= 2

# step 2: test odd 
for i in range(3, 1000, 2):
    while num%i == 0:
        num /= i
        if num == 1:
            print(i)

# largest factor is 7
```

```python
# POC trail division method - scaling

num = 13195

# step 1: make it odd
while num%2 == 0: # is dividable by 2
    num /= 2

# step 2: test odd 
for i in range(3, 1000, 2):
    while num%i == 0:
        num /= i
        if num == 1:
            print(i)

# largest factor is 29
```

```python
# answer trail division method - scaling more

num = 600851475143

# step 1: make it odd
while num%2 == 0: # is dividable by 2
    num /= 2

# step 2: test odd 
for i in range(3, num, 2): # start, end, step
    while num%i == 0:
        num /= i
        if num == 1:
            print(i)

# largest factor is 6857 (but jupyter notebook hangs cuz of the large while loop)
```

```python
# final answer trail division method - limiting upper end = faster

num = 600851475143

# step 1: make it odd
while num%2 == 0: # is dividable by 2
    num /= 2

# step 2: test odd 
limit = int(600851475143**0.5)
for i in range(3, limit+1, 2): # start, end, step
    while num%i == 0:
        num /= i
        if num == 1:
            print(i)
            
# largest factor is 6857
```
