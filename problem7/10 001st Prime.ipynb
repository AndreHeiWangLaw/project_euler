```python
# POC day 2

def get_nth_prime(nth):

    # dict of numbers and if prime or not
    numbers = {}
    for i in range(2, nth**2):
        numbers[i] = True

    count = 0
    for i in range(nth):
        for r in numbers:
            if numbers[r] == True:
                count += 1
                if count == nth: # check if nth
                    return r
                counter = 2
                while r*counter <= len(numbers)+1:
                    numbers[r*counter] = False
                    counter += 1
    
    
    return None


print(get_nth_prime(6))
```
