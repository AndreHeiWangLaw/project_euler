```python
"""

By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the th prime is 13.

What is the 10001 st prime number?

"""
```

```python
# POC 

def get_nth_prime(nth):

    # dict of numbers and if prime or not
    numbers = {}
    for i in range(2, nth**2):
        numbers[i] = True

    # p=2 and true 
    p = 2

    counter = 2
    print(len(numbers))
    while p*counter < len(numbers):
        print(p*counter)
        numbers[p*counter] = False
        counter += 1
            
        
    
    
    return True


print(get_nth_prime(6))
```
